```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
from scipy.stats import zscore
import sklearn
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.metrics import confusion_matrix
from xgboost import XGBClassifier

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV
from sklearn.naive_bayes import GaussianNB
from sklearn.utils import shuffle
from sklearn.neighbors import KNeighborsClassifier
```

```python
X = df4[['Сумма_выплат', 'Общая_сумма_покупки', 'Цена']]
y = df4['Кредит']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
clf = XGBClassifier()
clf.fit(X, y)

training_preds = clf.predict(X_train)
test_preds = clf.predict(X_test)

training_accuracy = accuracy_score(y_train, training_preds)
test_accuracy = accuracy_score(y_test, test_preds)

print('Training Accuracy: {:.4}%'.format(training_accuracy * 100))
print('Validation accuracy: {:.4}%'.format(test_accuracy * 100))
```

```python
df4['Кредит'].value_counts()
```

```python
df_loan = df4[df4['Кредит'] == 'Да']
df_not_loan = df4[df4['Кредит'] == 'Нет']
```

```python
df_loan = shuffle(df_loan)
df_not_loan = shuffle(df_not_loan)
```

```python
df_not_loan_493 = df_not_loan[:493]
```

```python
#put two df together so we get a df of two equally represented classes
twodf = [df_loan, df_not_loan_493]
edf = pd.concat(twodf, ignore_index=True)

#shuffle them again!
edf = shuffle(edf)
edf.head()
```

```python
edf.Кредит.value_counts()
```

```python
Xsh = edf[['Сумма_выплат', 'Общая_сумма_покупки', 'Цена']]
ysh = edf['Кредит']
```

```python
Xsh_train, Xsh_test, ysh_train, ysh_test = train_test_split(Xsh, ysh, test_size=0.2, random_state=1)
```

```python
clf = XGBClassifier()
clf.fit(Xsh, ysh)

training_preds_ = clf.predict(Xsh_train)
test_preds_ = clf.predict(Xsh_test)

training_accuracy_ = accuracy_score(ysh_train, training_preds_)
test_accuracy_ = accuracy_score(ysh_test, test_preds_)

print('Training Accuracy: {:.4}%'.format(training_accuracy_ * 100))
print('Validation accuracy: {:.4}%'.format(test_accuracy_ * 100))
```

```python
print(classification_report(ysh_test, test_preds_))
```
